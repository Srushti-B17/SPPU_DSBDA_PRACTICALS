# Detecting outliers using z-score

```python
import numpy as np
from scipy import stats
```

```python
z = np.abs(stats.zscore(df['Math_score']))
```

```python
print(z)
```

```python
threshold = 0.18
```

```python
sample_outliers = np.where(z <threshold)
sample_outliers
```

# Remove Outliners

```python
newdf=df
```

```python
for i in sample_outliers:
    newdf.drop(i,inplace=True)
```

```python
newdf
```
